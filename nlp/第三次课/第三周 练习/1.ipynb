```python
import nltk
from nltk.tokenize import sent_tokenize 
```

### 读取数据

```python
text_en = open(u'./data/text_en.txt',encoding='utf-8',errors='ignore').read()
```

### 切分语句

```python
tokens = sent_tokenize(text_en)
print(len(tokens))
```

### 分离缩略词实现切分词

```python
words = nltk.word_tokenize(text_en)
print(len(words))
```

### 通过分离标点实现切分词

```python
from nltk.tokenize import WordPunctTokenizer
tokenizer = WordPunctTokenizer()
words1 = tokenizer.tokenize(text_en)
print(len(words1))
```

### 提取词干

```python
from nltk import SnowballStemmer
print(SnowballStemmer.languages)
```

```python
english = SnowballStemmer('english')
for word in words:
    print('%s: %s' % (word, english.stem(word)))
```

### 去停用词

```python
from nltk.corpus import stopwords
stwords = stopwords.words('english')
print(stwords)
```

```python
print('before: %d' % len(words))
filtered = [word for word in words if word not in stwords]
print('after: %d' % len(filtered))
```

### 去标点符号

```python
from nltk.tokenize import RegexpTokenizer
import re
punct_pattern = re.compile("^['\"{}\\(\\)\\[\\]\\*&.?!,…:;]+$")
puncts_list = set([word for word in words if punct_pattern.match(word)])
print('original unique puncts count: %d' % len(puncts_list))

nopunct_tokenizer = RegexpTokenizer(r'\w+')
nopunct = nopunct_tokenizer.tokenize(text_en)
nopunct = [word for word in nopunct if word not in stwords]
print('no puncts count: %d' % len(nopunct))
for word in nopunct:
    if word in puncts_list:
        print("Punct found!")
```

### 低频词过滤

```python
from nltk.probability import FreqDist
fdist = FreqDist(nopunct)
```

```python
frequency = dict()
for word in nopunct:
    frequency[word] = frequency.get(word, 0) + 1
frequency = list(frequency.items())
frequency.sort(key=lambda x:x[1], reverse=True)
for word, count in frequency:
    print('%s: %s' % (word, count))
```

```python
threshold = 20
```

```python
print('Before filtering: %d' % len(nopunct))
filtered = [word for word in nopunct if fdist[word] > threshold]
print('After filtering: %d' % len(filtered))
```

### 绘制离散图，查看指定单词（Elizabeth, Darcy, Wickham, Bingley, Jane）在文中的分布位置

```python
from nltk.text import Text
```

```python
text = Text(filtered)
```

```python
to_plot = ['Elizabeth', 'Darcy', 'Wickham', 'Bingley', 'Jane']
text.dispersion_plot(to_plot)
```
