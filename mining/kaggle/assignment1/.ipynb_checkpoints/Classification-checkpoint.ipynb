```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cost(hx, y):
    return (1 / hx.shape[0]) * np.sum(-(y * np.log(hx) + (1 - y) * np.log(1 - hx)))
```

```python
def regression(x, target, h, iters, lr):
    samples, features = x.shape
    theta = np.zeros(shape=(features, 1))
    for i in range(iters):
        grad = np.zeros(shape=theta.shape)
        err = h(x, theta).reshape(samples, 1) - y
        for j in range(features):
            grad[j] = lr * np.sum(err * x[:, j].reshape(samples,1)) / samples
        theta -= grad
#         if i % 5000:
#             print(cost(h(x, theta), target))
    return theta
```

```python
def plot(X, Y):
    T = np.array([1 if e > 0.5 else 0 for e in Y])
    plt.scatter(X[:,0], X[:,1], c=T)
    plt.xlim(-1.5, 1.5)
    plt.ylim(-1.5, 1.5)
    plt.show()
```

```python
def min_max_scaling(x):
    min_val = np.min(x, axis=0)
    max_val = np.max(x, axis=0)
    res = np.zeros(shape=x.shape)
    for i in range(len(min_val)):
        res[:, i] = (x[:, i] - min_val[i]) / (max_val[i] - min_val[i])
    return res
```

```python
def z_score_scaling(x):
    res = np.zeros(shape=x.shape)
    mean_val = np.mean(x, axis=0)
    std_val = np.std(x, axis=0)
    for i in range(len(mean_val)):
        res[:, i] = (x[:, i] - mean_val[i]) / std_val[i]
    return res
```

```python
lr = 0.01
iters = 50000
h = lambda x, theta : 1/(1+np.exp(-x.dot(theta)))
a = np.array([[1, 0.2], [1, 0.5], [1, 0.6], [1, 0.46], [1, 0.78]])
y = np.array([[0], [1], [1], [0], [1]])
the = regression(a, y, h, iters, lr)
print(the)
h(a, the)
```

```python
a = np.random.randn(200, 2)
test = np.random.randn(200, 2)
y = np.array([1 if e[0] + e[1] > 0 else 0 for e in a])
plot(a, y)
```
