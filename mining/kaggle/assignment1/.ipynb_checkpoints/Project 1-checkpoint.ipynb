```python
import pandas as pd
import numpy as np
import kmeans
import regression
import preprocess
```

```python
data = pd.read_csv("data/raw data.csv")
```

```python
k = 2
```

```python
processed, _, _ = preprocess.z_score_scaling(data.values)
```

```python
result = kmeans.kmeans(processed, k, True)
```

```python
len(result)
```

```python
pd.DataFrame([[i+1, int(result[i])] for i in range(len(result))]).to_csv('result.csv', header=['Id', 'Predicted'], index=False)
```

```python
t = 0
for r in result:
    if r == 1:
        t += 1
```

```python
lr = 0.01
iters = 1000
the = regression.regression(processed, result.reshape(len(result), 1), regression.logistic_h, iters, lr, log=True)
```

```python
guess = regression.logistic_h(processed, the)
```

```python
guess = [1 if e > 0.5 else 0 for e in guess]
accurate = 0
for i in range(len(guess)):
    if guess[i] == result[i]:
        accurate+= 1
print(accurate / len(guess))
```
